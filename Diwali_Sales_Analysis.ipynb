```python
import numpy as np
import pandas as pd
import matplotlib as plt
%matplotlib inline
import seaborn as sns
```

```python
#import the csv file 
df = pd.read_csv('Diwali Sales Data.csv', encoding = 'unicode_escape')
```

```python
df.shape
```

```python
df.info()
```

```python
df.head()
```

```python
#drop unrelated/blank columns
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
df.info()
```

```python
#Find the null values and delete those values
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
#changes the amount datatype from float to int
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtype
```

```python
df.info()
```

```python
# describe() method returns description of the data in the DataFrame (i.e. count, mean, std, etc)
df.describe()
```

```python
# describe() method used for specific columns (Age,Orders,Amount)
df[['Age','Orders','Amount']].describe()
```

```python
#plotting barchart for Gender and counts

ag=sns.countplot(x='Gender',data=df)

for bars in ag.containers:
    ag.bar_label(bars)
    
# This barchar shows female's have done 7832 times shopping and men have done 3407 times shopping.
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
# Plotting barchart for Gender vs Amount

gm=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Gender',y='Amount',data=gm)

# The below barplot shows that female purchasing power is more then that of men and have done more shopping as compared to men.
```

```python
# Plotting barchart for Age Group and Count

ag=sns.countplot(data=df,x='Age Group',hue='Gender')

for bars in ag.containers:
    ag.bar_label(bars)
```

```python
# Plotting barchart for Age Group Vs Amount

aa=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Age Group',y='Amount',data=aa)

# From the above and below plot we can see that most of the buyers are female of 26-35 years.
```

```python
# plotting bar chart for states vs orders top 10

sales_state=df.groupby(['State'],as_index =False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(23,5)})
sns.barplot(x='State',y='Orders',data=sales_state)

#Uttar Pradesh has highest number of orders and Telangana has the lowest.
```

```python
# Plotting bar chart for state vs Amount top 10

Sale_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
                                                                                
sns.set(rc={'figure.figsize':(23,5)})
sns.barplot(x='State',y='Amount',data=Sale_state)

#Kerala is not in the list for top 10 states vs amount spend.
```

```python
# plotting bar chart Marital Status vs Count

ax = sns.countplot(data=df,x='Marital_Status')
for bars in ax.containers:
    ax.bar_label(bars)
sns.set(rc={'figure.figsize':(5,5)})

# Most of the shopping is done by Unmarried Women.
```

```python
# plotting bar chart for Marital Status vs Amount

ma=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Marital_Status',y='Amount',data=ma,hue='Gender')

# Below Graph shows Female have high purchasing power
```

```python
# Bar chart plot of different occupations vs count of purchase

oc=sns.countplot(data=df,x='Occupation')
for bars in oc.containers:
    oc.bar_label(bars)
sns.set(rc={'figure.figsize':(20,5)})

#Top three occupations with most products purchased are Healthcare,Govt and Automobile.
```

```python
# Top 10 Occupations with most amount spent on purchasing products.

pp=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
                                                                                
sns.set(rc={'figure.figsize':(23,5)})
sns.barplot(x='Occupation',y='Amount',data=pp)

# Top occupations are It sectors, healthcare etc.
```

```python
#Product Category

pc=sns.countplot(data=df,x='Product_Category')
for bars in pc.containers:
    pc.bar_label(bars)
sns.set(rc={'figure.figsize':(20,5)})

# Most purchasing are in clothing, food and electronics product category.
```

```python
# Amount spent on different product category.
sale_pc=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
                                                                                
sns.set(rc={'figure.figsize':(23,5)})
sns.barplot(x='Product_Category',y='Amount',data=sale_pc)

#Food being the highest and office being the lowest.
```

```python
# compare top 5 productid with there No. of Orders.
sale_as=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(5)
                                                                                
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_ID',y='Orders',data=sale_as)

# The below results shows the Top 5 productid with respect to there orders.
```

```python
# Conclusion
#Unmarried Women from age group 26-35 years from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation
#are more likely to buy products from food, clothing and electronic category.
```
